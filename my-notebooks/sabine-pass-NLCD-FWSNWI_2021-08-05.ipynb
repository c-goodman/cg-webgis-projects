Create an interactive map.

```python
import folium as fm
```

```python
# Creates Map and centers the map on Sabine Pass
mNLCD = leafmap.Map(location=[29.7452, -93.8833,], zoom_start=12)

mNLCD.add_basemap('NLCD 2016 CONUS Land Cover')
mNLCD.add_legend(builtin_legend='NLCD')
fm.Marker([29.7541, -93.8739], popup="Sabine Pass LNG").add_to(mNLCD)

mNLCD
```

```python
mNLCD.to_html("sabine-pass-NLCD_2021-08-05.html")
```
